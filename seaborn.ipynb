```python
import seaborn as sns
import matplotlib.pyplot as plt
import random as ra
```

```python
x=[1,3,4,2,4,6,5]
y=[10,11,14,15,13,12,28]
```

```python
sns.scatterplot(x=x,y=y)
plt.show()
```

```python
x_data=[ra.randint(i,i+5)for i in range(1,100)]
y_data=[ra.randint(i,i+100)for i in range(1,100)]
```

```python
sns.scatterplot(x=x_data,y=y_data)
plt.show()
```

```python
x_data=[ra.randint(i,i+5)for i in range(1,100)]
y_data=[ra.randint(i-10,i+10)for i in range(1,100)]
```

```python
sns.scatterplot(x=x_data,y=y_data)
plt.show()
```

```python
x_data=[ra.randint(i,i+5)for i in range(1,100)]
y_data=[ra.randint(i-10,i+10)for i in range(1,100)]
z_data=[ra.randint(1,3) for _ in range(len(x_data))]
```

```python
sns.scatterplot(x=x_data,y=y_data,hue=z_data)
plt.show()
```
